## Feature Exploration
### 1. Category of Alerts
Analyze the frequency of each alert category.

```python
# Count unique categories of alerts
df['Category'] = df['Category'].str.strip("[]'")  # Clean string format if needed
category_counts = df['Category'].value_counts()

# Plot the distribution of alert categories
plt.figure(figsize=(10, 6))
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.xticks(rotation=90)
plt.title('Alert Categories Frequency')
plt.ylabel('Count')
plt.xlabel('Alert Category')
plt.show()
```

### Insight:
This shows the prevalence of different types of alerts, such as reconnaissance or scanning events, which may help prioritize specific types for further alerting and policy enforcement.

### 2. Time-Based Analysis: DetectTime and EventTime
Explore how alerts occur over time to identify any patterns or anomalies.

```python
# Convert time columns to datetime format
df['DetectTime'] = pd.to_datetime(df['DetectTime'], errors='coerce')
df['EventTime'] = pd.to_datetime(df['EventTime'], errors='coerce')

# Check for any missing or NaT values after conversion
print(df[['DetectTime', 'EventTime']].isna().sum())

# Plot alerts over time
df.set_index('DetectTime')['Category'].resample('D').count().plot()
plt.title('Daily Alert Count Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Alerts')
plt.show()
```

### Insight:
This highlights any spikes or patterns in alert occurrences, which can indicate active threat periods or anomalies.

### 3. ByteCount and PacketCount Analysis
Analyze ByteCount and PacketCount for network usage trends and possible data exfiltration indicators.

```python
# Distribution plot for ByteCount and PacketCount
plt.figure(figsize=(7, 3))
plt.subplot(1, 2, 1)
sns.histplot(df['ByteCount'].dropna(), kde=True)
plt.title('ByteCount Distribution')

plt.subplot(1, 2, 2)
sns.histplot(df['PacketCount'].dropna(), kde=True)
plt.title('PacketCount Distribution')

plt.show()
```

### Insight:

## Source and Target Analysis
#### 4. Source and Target IP Patterns
Analyze source and target patterns to identify frequently targeted or attacking IPs.

```python
import ast  # For safely evaluating strings that represent dictionaries

# Extract IPs from Source and Target columns with a try-except approach to handle missing or inconsistent keys
def extract_ip(data, key='IP4'):
    try:
        # Parse the string to a dictionary safely
        data_dict = ast.literal_eval(data)
        # Extract the first IP if the key exists
        return data_dict[0].get(key, [None])[0] if isinstance(data_dict, list) and data_dict else None
    except (ValueError, KeyError, TypeError):
        return None

# Apply the function to Source and Target columns
df['Source_IP'] = df['Source'].apply(lambda x: extract_ip(x))
df['Target_IP'] = df['Target'].apply(lambda x: extract_ip(x))

# Verify the results
print("Top Source IPs:\n", df['Source_IP'].value_counts().head(10))
print("\nTop Target IPs:\n", df['Target_IP'].value_counts().head(10))
```

### Insight: 
This helps identify the most frequent attackers or targets, which could be the focus for specific policy enforcement rules.
